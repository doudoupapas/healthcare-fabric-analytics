#### **# Read Excel from Lakehouse Files**

```python
import pandas as pd
pdf = pd.read_excel("abfss://HealthcareAnalytics@onelake.dfs.fabric.microsoft.com/HealthcareLakehouse.Lakehouse/Files/bronze/drugs/drug_inventory.xlsx")

# Convert to Spark DataFrame
df_drugs = spark.createDataFrame(pdf)

print(f"Row count: {df_drugs.count()}")
display(df_drugs.limit(5))
```

#### **# Save to Bronze Delta table**

```python
df_drugs.write.format("delta").mode("overwrite").saveAsTable("bronze_drug_inventory")
print("✅ bronze_drug_inventory saved")
```
